```python
K = 1
P = np.random.normal(size=(rm_small.shape[0], K))
Q = np.random.normal(size=(rm_small.shape[1], K))
```

```python
def matrix_factorization(R, P, Q, K, steps=10000, alpha=0.0001, beta=0.002):
    Q = Q.T
    for step in range(steps):
        for i in range(len(R)):
            for j in range(len(R[i])):
                if R[i][j] == 0:
                    continue
                eij = R[i][j] - np.dot(P[i,:],Q[:,j])
                for k in range(K):
                    x = P[i][k]
                    P[i][k] += alpha * ((2 * eij * Q[k][j])  - beta * P[i][k])
                    Q[k][j] += alpha * ((2 * eij * x)  - beta * Q[k][j])
    return P, Q.T
```

```python
P_, Q_ =  matrix_factorization(rm_small.values.copy(), P.copy(), Q.copy(), 1)
```

```python
(rm_small>0).sum().sum()/(rm_small.shape[0]*rm_small.shape[1])
```

```python
from sklearn.metrics import mean_squared_error as mse
rm_ = np.dot(P_, Q_.T)
mse(rm_small.values[rm_small > 0], rm_[rm_small > 0])**0.5
```
